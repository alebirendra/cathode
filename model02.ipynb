```python
import numpy as np
from itertools import permutations
from itertools import combinations
# Get all permutations of [1, 1, 3]
perm = combinations(['Mn', 'Mn', 'Mg', 'Bi', 'Co', 'Cu', 'Zn', 'Ni', 'Ca', 'Ti', 'V', 'Sc', 'Cr', 'Al',
                    'Mg', 'Bi', 'Co', 'Cu', 'Zn', 'Ni', 'Ca', 'Ti', 'V', 'Sc', 'Cr', 'Al'],6)
#perm = permutations(['Mn', 'Mn', 'Mg', 'Bi', 'Co', 'Cu', 'Zn', 'Ni', 'Ca', 'Ti', 'V', 'Sc', 'Cr'],5)

lst2=[]
for i in list(perm):
    lst2.append(i)
```

```python
lst2[0]
```

```python
lst=[]
for a in lst2:
    if a not in lst:
        lst.append(a)
        
len(lst)
```

```python
lst=[]
for i in range(0, len(lst2)):
    if a not in lst:
        lst.append(a)
        
len(lst)
```

```python
lst[0]
```

```python


elementsg = []
voltageavg =[]
deltaE1g = []
deltaE2g = []
a1g = []
a2g = []
delta_ag = []
c1g = []
c2g = []
delta_cg = []
v1g = []
v2g = []
delta_vg = []


for i in range(0,len(lst)):
    myelem = ''
    v_av = 0
    e1g_av = 0
    e2g_av = 0

    a1_av=0
    a2_av=0
    da_av=0

    c1_av=0
    c2_av=0
    dc_av=0
    
    v1_av=0
    v2_av=0
    dv_av=0
    
    for j in lst[i]:
        myelem += ''+j

  #  dfg['elementsg'][i] = myelem
    elementsg.append(myelem)  
    for j in range(0,len(lst[i])):

        v_av = v_av+float(df.loc[df['elements']==str(lst[i][j])]['voltageav'])
        e1g_av = e1g_av+float(df.loc[df['elements']==str(lst[i][j])]['deltaE1'])
        e2g_av = e2g_av+float(df.loc[df['elements']==str(lst[i][j])]['deltaE2'])
        a1_av = a1_av+float(df.loc[df['elements']==str(lst[i][j])]['a1'])
        a2_av = a2_av+float(df.loc[df['elements']==str(lst[i][j])]['a2'])
        da_av = da_av+float(df.loc[df['elements']==str(lst[i][j])]['delta_a(%)'])

        c1_av = c1_av+float(df.loc[df['elements']==str(lst[i][j])]['c1'])
        c2_av = c2_av+float(df.loc[df['elements']==str(lst[i][j])]['c2'])
        dc_av = dc_av+float(df.loc[df['elements']==str(lst[i][j])]['delta_c(%)'])

        v1_av = v1_av+float(df.loc[df['elements']==str(lst[i][j])]['v1'])
        v2_av = v2_av+float(df.loc[df['elements']==str(lst[i][j])]['v2'])
        dv_av = dv_av+float(df.loc[df['elements']==str(lst[i][j])]['delta_v(%)'])

        

    voltageavg.append(v_av/len(lst[0]))  
    deltaE1g.append(e1g_av/len(lst[0]))  
    deltaE2g.append(e2g_av/len(lst[0]))  

    a1g.append(a1_av/len(lst[0]))  
    a2g.append(a2_av/len(lst[0]))  
    delta_ag.append(da_av/len(lst[0]))  

    c1g.append(c1_av/len(lst[0]))  
    c2g.append(c2_av/len(lst[0]))  
    delta_cg.append(dc_av/len(lst[0]))  

    
    v1g.append(v1_av/len(lst[0]))  
    v2g.append(v2_av/len(lst[0]))  
    delta_vg.append(dv_av/len(lst[0]))  

    

        
listgen = list(zip(elementsg, voltageavg, deltaE1g, deltaE2g, a1g, a2g, delta_ag, c1g, c2g, delta_cg, v1g, v2g, delta_vg))
 
dfg6 = pd.DataFrame(listgen,
                  columns=['elementsg', 'voltageavg', 'deltaE1g', 'deltaE2g', 'a1g', 'a2g', 'delta_ag', 'c1g','c2g',
                           'delta_cg', 'v1g', 'v2g', 'delta_vg'])
       
    
```

```python
dfg6.to_csv('M6_data_gen.csv', index=None)
```
