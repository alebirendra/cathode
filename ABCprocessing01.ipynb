```python
df=pd.read_csv('dataABC.dat',sep='\s+|\s+', header=None)
```

```python
df.rename(columns={0: 'structure', 1: 'steps', 2: 'formula', 3: 'a', 4: 'b', 5: 'c', 6: 'alpha', 7: 'beta', 8: 'gamma', 9: 'volume', 10: 'energy(Ry)'}, inplace=True)
```

```python
df
```

```python
df1 = df[df['steps']==1]
df2 = df[df['steps']==2]
df3 = df[df['steps']==3]

df1 = df1.drop(['b', 'alpha','beta','gamma'],axis=1)
df1.rename(columns={ 'volume': 'v', 'energy(Ry)': 'e'}, inplace=True)

df2 = df2.drop(['b', 'alpha','beta','gamma'],axis=1)
df2.rename(columns={ 'a': 'a1', 'c': 'c1', 'volume': 'v1', 'energy(Ry)': 'e1'}, inplace=True)

df3 = df3.drop(['b', 'alpha','beta','gamma'],axis=1)
df3.rename(columns={ 'a': 'a3', 'c': 'c3', 'volume': 'v3', 'energy(Ry)': 'e3'}, inplace=True)
```

```python
df1['a1']=df2['a1'].values
df1['a3']=df3['a3'].values
df1['c1']=df2['c1'].values
df1['c3']=df3['c3'].values
df1['v1']=df2['v1'].values
df1['v3']=df3['v3'].values
df1['e1']=df2['e1'].values
df1['e3']=df3['e3'].values
```

```python
import re


elements=[]
for i in range(0,len(df1['formula'])):
    lst=[]

    my_str = str(df1['formula'].values[i][:-6])
    my_list = re.split(r'(\d+)', my_str)

    str1= str(df1['formula'].values[i][3:5])
    str2= str(df1['formula'].values[i][5:7])
    str3= str(df1['formula'].values[i][7:-6])

    lst.append(str1)
    lst.append(str2)
    lst.append(str3)

    
    elements.append(lst)

```

```python
df1.rename(columns={'structure':'elements'}, inplace=True)
df1['elements'] = elements
df1=df1.drop(['steps'],axis=1)
```

```python

Li   =    -129.34419679/9
o = -256.511036/8
ry=13.6057
deltaE1=[]
deltaE3=[]
df1['volt1'] = ry*(df1['e'].values- df1['e1'].values - 1*Li)/1
df1['volt_av'] = ry*(df1['e'].values- df1['e3'].values - 3*Li)/3

for i in range(0, len(df1['elements'])):  #   len(df1['elements'])
    esum=0
    for j in range(0, len(df1['elements'].values[i])):
       # print(df1['elements'].values[i][j])
        
       # print(i)
       # print(j)

        esum += (ele.loc[ele['element']== str((df1['elements'].values[i][j]))  ]['Eper_atom'].values)[0]  

  #  print(df1['e'].values[i])
    de1=ry*(df1['e'].values[i] - esum - 3*Li-6*o)/12
    de3=ry*(df1['e3'].values[i] - esum -6*o)/9

    deltaE1.append(de1)
    deltaE3.append(de3)

df1['deltaE1'] = deltaE1
df1['deltaE3'] = deltaE3
```
